```python
import os
import cv2
import numpy as np

def load_images_from_folder(folder):
    images = []
    labels = []
    for class_folder in os.listdir(folder):
        class_path = os.path.join(folder, class_folder)
        if os.path.isdir(class_path):
            for filename in os.listdir(class_path):
                img_path = os.path.join(class_path, filename)
                img = cv2.imread(img_path)
                img = cv2.resize(img, (224, 224))  # Resize images
                if img is not None:
                    images.append(img)
                    labels.append(class_folder)
    return np.array(images), np.array(labels)

data_folder = "./Traffic Sign/DATA"
images, labels = load_images_from_folder(data_folder)
```

```python
def extract_low_level_features(images):
    features = []
    for img in images:
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        # Apply your low-level feature extraction techniques (e.g., edges, textures)
        # For example, you can use Canny edge detection:
        edges = cv2.Canny(gray, 100, 200)
        features.append(edges.flatten())
    return np.array(features)

low_level_features = extract_low_level_features(images)
```

```python
import tensorflow as tf
from tensorflow.keras.applications.vgg16 import VGG16
from tensorflow.keras.applications.vgg16 import preprocess_input

def extract_high_level_features(images):
    model = VGG16(weights='imagenet', include_top=False)
    features = []
    for img in images:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # VGG expects RGB images
        img = cv2.resize(img, (224, 224))
        img = preprocess_input(img)
        feature = model.predict(np.expand_dims(img, axis=0))
        features.append(feature.flatten())
    return np.array(features)

high_level_features = extract_high_level_features(images)
```
